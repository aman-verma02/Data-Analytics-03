```python
import pandas as pd
df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.info()
```

```python
df.describe()
```

```python
df.describe(include='all')
```

```python
# checking missing values
df.isnull().sum()
```

```python
df.columns= df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns = {'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.head()
```

```python
# create a new column for age_grp
labels = ['Young Adult', 'Adult', 'Middle Aged', 'Senior']
bins = [0, 25, 40, 60, 100]
df['age_grp'] = pd.cut(df['age'], bins, labels=labels)
df.head()
```

```python
df[['age', 'age_grp']].head()
```

```python
# create column purchase_frequency_days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Yearly': 365,
    'Quarterly': 90,
    'Bi-Weekly': 21,
    'Annually': 365 ,
    'Every 3 Months': 90,
    'Every 6 Months': 180
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases', 'purchase_frequency_days']].head()
```

```python
df[['discount_applied', 'promo_code_used']].head()
```

```python
# check whether discount applid is equal to promo code used
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# Deleting the promo code used column
df  = df.drop('promo_code_used', axis=1)
```

```python
df.columns
```

```python
df.to_csv('customer_shopping_behavior_dataset.csv', index = False)
```

```python
# Install required libraries

from sqlalchemy import create_engine
from urllib.parse import quote_plus

# SQL Server connection
username = "sa"
password = "your_password"
host = "localhost"
port = "1433"
database = "customer_behavior"

# Note: requires Microsoft ODBC Driver installed separately on your machine
driver = quote_plus("ODBC Driver 17 for SQL Server")
engine = create_engine(f"mssql+pyodbc://{username}:{password}@{host},{port}/{database}?driver={driver}")

# Write DataFrame to SQL Server
df.to_sql("customer", engine, if_exists="replace", index=False)

# Read back sample (SQL Server uses TOP instead of LIMIT)
pd.read_sql("SELECT TOP 5 * FROM customer;", engine)
```
